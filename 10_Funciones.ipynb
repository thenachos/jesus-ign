## FUNCIONES

Una función es un bloque reutilizable de código o declaraciones de programación diseñadas para realizar una determinada tarea.

Para declarar una función en Python deberemos usar la palabra clave def. 
El bloque de función de código se ejecuta solo si se llama o invoca la función.



```python
#sintaxis
def function_name():
  code
  code

# funciones sin parametros
def adicion():
  a = 2
  b = 3
  total = a + b
  print(total)

adicion()
```


# RETORNAR VALORES DE UNA FUNCIÓN
para retornar un valor de una función nosotros usamos la palabra return, seguida de la varibale que vamos a retornar.

Si no lo hace su valor final sera **"None"**

```python
print(adicion())

def adicion():
  a = 2
  b = 3
  total = a + b
  return total
```
