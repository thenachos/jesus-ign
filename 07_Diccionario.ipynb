##DICCIONARIO

Es una colección desordenada, modificables y emparejados.

```python
my_dict = dict()
other_dict = {}
print(type(my_dict))
print(type(other_dict))
```

Una de las características más importantes de los diccionarios es que están emparejados, lo cual significa que siempre se van a componer de una llave y un valor.

key:value

```python
person = {
    'name' : 'Abdiel',
    'last_name' : 'Hernández',
    'Age' : 19,
    'skills' : ['Javascript', 'react', 'Python', 'switf'],
    'address' : {
        'street' : 'Fco. I. Madero',
        'teléfono' : 7821004745
    }
}
```

##CALCULAR LONGITUD

```python
print(len(person))
```

##ACCEDER A LOS ELEMENTOS

```python
print(person['name'])
print(type(person['name']))

print(person['last_name'])
print(type(person['last_name']))

print(person['skills'])
print(type(person['skills']))

print(person['address'])
print(type(person['address']))
```

```python
print(person.get('name'))
print(person.get('country'))
print(person.get('skills'))
print(person.get('city'))
```

##AGREGAR NUEVOS CAMPOS

```python
print(person)
person['country'] = 'Poza Rica'
print(person)
```

##MODIFICAR NUEVOS CAMPOS

```python
print(person)
person['last_name'] = 'Martínez'
print(person)
```

##VERIFICAR LLAVES

```python
print('name' in person)
print('marital_status' in person)
```

##ANALIZAR TODAS LAS LLAVES EXISTENTES

```python
print(person.keys())
```

##ANALIZAR TODOS LOS VALORES EXISTENTES

```python
print(person.values())
```
