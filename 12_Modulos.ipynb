## Módulo para trabajar con operaciones matemáticas

```python
import math
print(math.pi)
print(math.sqrt(16))
print(math.pow(2,3))
print(math.floor(9.84))
```

```python
# Podemos importar funciones especificas de cada módulo 
from math import pi, sqrt, pow
print(pi)
print(sqrt(16))
print(pow(2,3))
print(math.floor(9.84))
```

```python
# También podemos importar todas las funciones de jalón.
print(pi)
print(sqrt(16))
print(pow(2,3))
print(math.floor(9.84))

# También podemos renombrar una función
from math import pi as NumPi
print(NumPi)
```
