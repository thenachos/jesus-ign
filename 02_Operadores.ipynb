##OPERADORES ARITMETICOS

```python
print('Suma:', 1 + 2)
print('Resta:', 2 - 1)
print('Multiplicación:', 2 * 3)
print('División:', 4 / 2) #La división en Python asigna automáticamente números flotantes
print('División:', 7 / 2)
print('División sin decimales:', 7 // 2)
print('Módulo:', 3 % 2)
print('Exponente:', 2 ** 3)
```

<b>Podemos también unir palabras haciendo uso de operadores</b>

```python
print('Hola', 'Python')
print('Hola' + 'Python')
print('Hola' + str(5))
print('Hola ' * 5)
```

```python
##Estas funciones nos facilitan mucho ciertos trabajos, como por ejemplo:
for i in range(11):
  print("*" * i)
```

##OPERADORES DE COMPARACIÓN

```python
print('Igual a:', 2 == 2)
print('Igual a:', 2 == 2.00)
print('Igual a:', 2 == 1.99999)
```

```python
print('Distinto de:', 2 != 2)
print('Distinto de:', 2 != 3)
```

```python
print('Mayor que:', 2 > 3)
print('Menor que:', 2 < 3)
print('Mayor o igual que:', 4 >= 3)
print('Mayor o igual que:', 3 >= 3)
print('Menor o igual que:', 3 <= 4)
print('Menor o igual que:', 4 <= 4)
```

<b>También podemos hacer uso de estos operadores junto a texto</b>

```python
print(len('Mango') == len('Melón'))
print(len('Mango') != len('Melón'))
print(len('Mango') < len('Melón'))
print(len('Mango') > len('Melón'))
```

##OPERADORES DE COMPARACIÓN CON TEXTO

* is Regresa <b>True</b> si ambas variables son del mismo tipo.
* is not Regresa <b>True</b> si ambas variables <b>NO</b> son del mismo tipo.
* in Regresa <b>True</b> si el valor inicial se encuentra en el valor secundario.
* not in Regresa <b>True</b> si el valor inicial <b>NO</b> se encuentra en el valor secundario.





```python
print('1 is 1', 1 is 1)                        # True - El tipo de dato es el mismo.
print('1 is not 2', 1 is not 2)                # True - Aunque son el mismo tipo de dato no son de valor similar.
print('A in Abdiel', 'A' in 'Abdiel')          # True - 'A' Se encuentra en el String.
print('P in Abdiel', 'A' in 'Abdiel')          # False - No se encuentra una 'L' Mayúscula.
print('coding', 'coding' in 'coding for all'). # True - Porque la palabra se encuentra en la oración.
print('a in an:', 'a' in 'an')                 # True - Porque la letra se encuentra en la oración.
print('4 is 2 ** 2:', 4 is 2 ** 2)             # True - Porque hay el mismo valor de ambos lados.
```

##OPERADORES LÓGICOS

* and Regresa True si ambas condiciones se cumplen
* or Regresa True si una o ambas condiciones se cumplen
* not Regresa True si el valor es Falso

```python
print( 3 > 2 and 4 > 3 ) # True - Porque ambas condiciones se cumplen.
print( 3 > 2 and 4 < 3 ) # True - Porque la segunda condición es falsa.
print( 3 < 2 and 4 < 3 ) # True - Porque ambas condiciones son falsas.
print('True and True', True and True)

print( 3 > 2 or 4 > 3 ) # True - Porque ambas condiciones se cumplen.
print( 3 > 2 or 4 < 3 ) # True - Porque una de las condiciones se cumplen.
print( 3 < 2 or 4 < 3 ) # False - Porque ambas condiciones son falsas.
print('True or False', True or False)

print( not 3 > 2 ) # False - La afirmación es verdadera, entonces regresa lo contrario.
print( not True )  # False - Negación, el operador muestra el contrario.
print( not False ) # True
print( not not True ) # True
print( not not False ) # False
```

##ACTIVIDAD
Crea un programa que convierta la temperatura de grados Fahrenheit a Celcius.

```python
fahrenheit = float (input("Ingrese grados Fahrenheit: "))
celcius = ( fahrenheit - 32 ) / 1.8
print(f'Grados Celcius: {celcius}')
```
