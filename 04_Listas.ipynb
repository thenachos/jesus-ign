##LISTAS

Dentro de programación existen distuntas maneras de almacenar la información:

* Tuple
* Set
* Dictionay
* Lists

La última la analizaremos en esta ocasión.

Una lista es una colección de diferentes tipos de datos que tiene un orden modificable. Una lista puede estar vacía o contenewr distintos tipos de items.

Si esta definición no les suena o les parece un poco complicada, tal vez sea más fácil compararlas con lo que anteriormente vimos como Array.

```python
my_list = [1, 2, 3, 4, 5]
print(my_list)
print(len(my_list))

other_list = [1, 2.5, 3, 'Abdiel']
print(other_list)                   #Muestra los elementos de la lista.
print(len(other_list))              #Muestra la cantidad de caracteres de la lista.
```

```python
my_list = [4, 5, 6, 7, 8, 9]
print(my_list)                #Imprime los elementos de la lista.
print(type(my_list))          #Muestra el tipo de datos que contiene la lista.
print(len(my_list))           #Muestra la cantidad de caracteres de la lista.
print(my_list[0])             #Imprime el elemento de la primera posición de la lista.
```

```python
other_list = [1, 2.5, 3, 'Abdiel', 'Hernández']
print(other_list)
print(type(other_list))                           #Al contener varios tipos de caracter este muestra que es de tipo 'list'. 
```

```python
print(other_list[0])
print(other_list[-1])
```

##CONTAR COMPONENTES DE UNA LISTA

```python
other_list = [1, 2.5, 3, 'Abdiel', 'Hernández']
print(other_list.count("Hernández"))
print(other_list.count("Abdiel"))
```

##AGREGAR COMPONENTES A UNA LISTA

```python
frutas = ['Banana', 'Manzana', 'Sandía', 'Limón',]
print(frutas)
frutas.append('Manzana')                            #Agrega la palabra 'Manzana' en la última posición.
print(frutas)
frutas.append('Aguacate')                           #Agrega la palabra 'Aguacate' en la última posición.
print(frutas)
frutas.append(7.5)                                  #También puede agregar elementos de otro tipo a la lista.
print(frutas)
```

##INSERTAR COMPONENTES A UNA LISTA

```python
frutas = ['Banana', 'Manzana', 'Sandía', 'Limón',]
print(frutas)
frutas.insert(2, 'Melón')                           #Inserta la palabra 'Melón' en la posición 3.
print(frutas)
frutas.insert(3, 'Naranja')                         #Insertar la palabra 'Naranja' en la posición 4.
print(frutas)
```

##REMOVER COMPONENTES A UNA LISTA

```python
frutas = ['Banana', 'Manzana', 'Sandía', 'Limón',]
print(frutas)
frutas.remove('Limón')                           #Elimina la palabra 'Limón' de la lista.
print(frutas)
frutas.remove('Banana')                          #Elimina la palabra 'Banana' de la lista.
print(frutas)
```

##UNIR LISTAS

```python
positivos = [1, 2, 3, 4]
cero = [0]
negativos = [-1, -2, -3, -4]

enteros = positivos + cero + negativos   #Une las tres listas en una.

print(enteros)                           #Imprime la lista unificada.
print(len(enteros))
```
